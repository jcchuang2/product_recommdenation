```python
import pandas as pd
import numpy as np
```

```python
amazon = pd.read_csv('amazon.csv')
brands = pd.read_csv('brands.csv').dropna(subset=['brand', 'release_date'])
```

```python
df = pd.merge(amazon, brands[['url', 'brand', 'release_date']], left_on='product_link', right_on='url', how='inner')
```

```python
print(amazon.shape)
print(df.shape)
```

```python
df.columns
```

```python
df = df.drop(columns=['url', 'user_name', 'review_id', 'review_title', 'review_content'])
```

```python
df.head()
```

```python
df['user_id'] = df['user_id'].str.split(',')
df = df.explode('user_id', ignore_index=True)
```

```python
df.head()
```

```python
# Create a product catalog containing one row per unique product_id.
# We assume each product has consistent details.
product_catalog = df.drop_duplicates(subset='product_id').set_index('product_id')

# Container for additional rows to be appended.
additional_rows = []

# Group by user_id to check how many unique products each user already has.
for user, group in df.groupby('user_id'):
    # Set of product_ids already purchased by the user.
    purchased_products = set(group['product_id'])
    num_products = len(purchased_products)
    
    # Determine how many additional products are needed for this user.
    needed = max(0, 5 - num_products)
    
    if needed > 0:
        # List available product_ids that the user has not yet purchased.
        available_products = list(set(product_catalog.index) - purchased_products)
        
        # If available products are fewer than needed, sample with replacement.
        if len(available_products) < needed:
            sampled_products = np.random.choice(available_products, size=needed, replace=True)
        else:
            sampled_products = np.random.choice(available_products, size=needed, replace=False)
        
        # For each sampled product, get the full product details and assign the current user_id.
        for product in sampled_products:
            product_details = product_catalog.loc[product].to_dict()
            product_details['user_id'] = user  # Update the user_id for this new row.
            # Ensure the product_id is set (in case it was removed by setting it as the index).
            product_details['product_id'] = product
            additional_rows.append(product_details)

# Create a DataFrame from the additional rows, if any.
if additional_rows:
    additional_df = pd.DataFrame(additional_rows)
    # Combine the original DataFrame with the additional rows.
    final_df = pd.concat([df, additional_df], ignore_index=True)
else:
    final_df = df.copy()

final_df = final_df.sort_values(by='user_id').reset_index(drop='True')
final_df.head()
```

```python
print('Dataframe shape:', final_df.shape)
print('Number of unique users:', len(final_df['user_id'].unique()))
print('Number of unique products:', len(final_df['product_id'].unique()))
```

```python
final_df.to_csv('transactions_brand.csv', index=False)
```
